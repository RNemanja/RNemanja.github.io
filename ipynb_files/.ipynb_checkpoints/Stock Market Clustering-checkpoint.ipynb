# Sample solution

We'll analyse clustering of publicly-traded companies based on the similarity of their stock returns over time. Let us first read the data from the provided csv files.

We'll also try to highlight a few quick/easy ways to do things in pandas/numpy along the way that you may fight useful in the future.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

##### Read company names into a dictionary
def readNamesIntoDict():
    d = dict()
    input_file = csv.DictReader(open("SP_500_firms.csv"))
    # Firm list obtained from https://github.com/datasets/s-and-p-500-companies
    for row in input_file:
        #print(row)
        d[row['Symbol']] = [row['Name'],row['Sector']]
    return d

namesDict = readNamesIntoDict()
compNames = namesDict.keys()

##### Read price data with pandas 
filename = 'SP_500_close_2015.csv'
priceData = pd.read_csv(filename,index_col = 0)
```

## Main part

### Stock returns 

We'll start by calculating the stock returns. This is possible to do in several ways, as the below methods demonstrate.

```python
# Pandas makes this easy
returns = priceData.pct_change() # This calculates all returns
returns = priceData / priceData.shift(1) - 1 # This does the same

# We could also loop through the data
def compReturn(symbol,prices):
    # A standard loop through the prices that accomplishes the same for a company with the desired symbol 
    compPrice = prices[symbol]
    compRet = []
    for i in range(1,len(compPrice)):
        compRet.append((compPrice[i]-compPrice[i-1])/compPrice[i-1])
    return compRet

#compReturn('AAPL',priceData) #... repeat for all companies
```

From the returns, we can find the highest and lowest daily and yearly returns, and the highest and lowest yearly volatilities. Here's code that would avoid writing loops.

```python
print('Highest and lowest daily returns\n')

# Highest returns - we could loop through the data but here's an alternative way
print(returns.stack().nlargest(3))
# Lowest returns
print(returns.stack().nsmallest(3))


print('\nHighest and lowest yearly returns\n')

# Calculate total returns
#totalRets = priceData.ix[-1,:]/priceData.ix[0,:]-1 # deprecated
totalRets = priceData.iloc[-1,:]/priceData.iloc[0,:]-1
# Highest returns
print(totalRets.nlargest(3))
# Lowest returns
print(totalRets.nsmallest(3))


print('\nHighest and lowest volatilities\n')

# Calculate volatilities
vols = returns.std()
# Highest volatilities
print(vols.nlargest(3))
# Lowest volatilities
print(vols.nsmallest(3))
```

The highest and lowest daily returns during 2015 were recorded for FCX (Freeport-McMoRan on the day when the activist investor Carl Icahn decided to take interest...) and PWR (Quanta Services Inc on disappointing quarterly results). 

Overall the companies that did best and worst over the year were Netflix and Chesapeake energy. 

The most and least volatile stocks were Freeport-McMoran and Coca-Cola company.

### Correlations
Next we'll turn to correlations between the companies daily returns. We can calculate these either directly from the definition or using the pandas built-in function.

```python
correl = returns.corr() # Easy using pandas

def calcCorrelation(returns,company1,company2):
    # Calculates correlations from the definition
    
    rets1 = returns[company1]
    rets2 = returns[company2]

    rets1Sq = rets1 * rets1
    rets2Sq = rets2 * rets2
    
    pairwise =  rets1 * rets2
    
    n = len(rets1)-1
    
    corrs = (pairwise.sum()*n - rets1.sum()*rets2.sum())/(np.sqrt(n*rets1Sq.sum()-rets1.sum()*rets1.sum())*np.sqrt(n*rets2Sq.sum()-rets2.sum()*rets2.sum()))
    return corrs

print(calcCorrelation(returns,'AAPL','AMZN'))
print(correl['AAPL']['AMZN'])    
```

Now let's analyse the top correlations for tech companies.

```python
def printClosestCorrelations(correlations,companies,companyNames,nCorrs=10):
    # Prints the (nCorrs) highest correlated companies for each company in the list companies
    for comp in companies:
        highestCorrs = correlations[comp].nlargest(nCorrs+1) # get nCorrs+1 highest correlated companies (including itself)
        highestCorrs = highestCorrs.to_frame('Correlation')
        highestCorrs['Name'] = [companyNames[s][0] for s in highestCorrs.index] # get the names of the correlated companies 
        print('\n')    
        print(companyNames[comp][0])
        print('\n')
        print(highestCorrs[1:]) # ignore correlation with self
        

companiesToTest = ['AMZN','MSFT','GOOGL','AAPL','FB']

printClosestCorrelations(correl,companiesToTest,namesDict)     
```

### Clustering 

Let us next implement the clustering algorithm. We first need to get a list of the correlations. Here are a couple of alternative ways to get it.

```python
def createCorrelationListSlow(correl):
    # Returns a list of (unique, non-self) correlations from a pandas dataframe 
    L = [] # create list
    for index,company in enumerate(correl.columns):
        for secondCompany in correl.columns[index+1:]:
            L.append((correl[company][secondCompany],company,secondCompany))
    return L


def createCorrelationList(correl):
    # Returns a list of (unique, non-self) correlations from a pandas dataframe via correlation matrix
    # As above but faster
    nComp = len(correl.columns)
    compNames = list(correl.columns)

    correlMat = correl.as_matrix()
    L = [] # create list
    for i in range(nComp):
        for j in range(i+1,nComp):
            L.append((correlMat[i,j],compNames[i],compNames[j]))
    return L

edges = createCorrelationList(correl)

# For fun, let's do it in numpy too.

def createCorrelationListNumpy(correl):
    # The idea is to get the weight, source, and destination as arrays and then zip them together as a list
    # This is a bit more code but considerably faster 
    
    # First let's get the desired correlations as a 1D array 
    dataM = correl.as_matrix() # get just correlations (weights) as matrix
    inds2D = np.triu_indices(len(correl.columns),k=1) # Upper diagonal indices, k=1 ignores diagonal itself
    w = dataM[inds2D] # weights ie correlations
    
    # Next we need the sources and the destinations for each correlation
    # First let's get the companies as an array, and their count
    nComp = len(correl.index)
    lComp = np.array(correl.index)
    
    inds1D = np.ravel_multi_index(inds2D,(nComp,nComp)) # The upper diagonal indices of 2D matrix "raveled" into 1D array indices
    
    s = np.repeat(lComp,nComp) # Source as 1D array (repeat companies n times)
    s = s[inds1D] # sources
    d = np.tile(lComp,nComp) # Destination as 1D array (tile company list n times)
    d = d[inds1D] # destinations

    # Zip and list - tuples of form (correlation,source,destination)
    return list(zip(w,s,d)) 

edges = createCorrelationListNumpy(correl)

#%timeit createCorrelationListSlow(correl)
#%timeit createCorrelationList(correl)
#%timeit createCorrelationListNumpy(correl)

import pickle
file_name = 'cluster_correlations'

with open(file_name, 'wb') as f:
    pickle.dump(correl, f)
    pickle.dump(edges, f)
```

Using the edge list we can implement the clustering algorithm. We won't repeat the algorithm steps here, but the idea is to loop through the highest $k$ correlations and create clusters based on these connections between the firms. The main algorithm is in the function `kruskalCluster` that uses the two other helper functions for clarity.

```python
def findBottom(node,nextNodes):
    # Recursive helper function to get to the "bottom" of a set/cluster starting from node in dictionary nextNodes
    # Start from node and loop through dictionary nextNodes until find node points to itself 
    bottomNode = nextNodes[node]
    if bottomNode != node:
        bottomNode = findBottom(bottomNode,nextNodes)
    return bottomNode

def mergeSets(node1,node2,nextNodes,setStarters):
    # Helper function to merge two sets containing node1, node2 using the dictionary nextNodes
    # Find the bottom in both sets and make one bottom node point to the other
    bottom1 = findBottom(node1,nextNodes)
    bottom2 = findBottom(node2,nextNodes)
    nextNodes[bottom1] = bottom2
    try: # no longer a set starter
        setStarters.remove(bottom2)
    except: 
        pass
    return nextNodes

def kruskalCluster(edgeList, correl, k=200):
    """
     A version of Kruskal's algorithm for stock market clustering
     "Naive" implementation (not optimised for efficiency)
    
     Parameters: 
         edgeList - list of edges of the form (weight,source,destination)
         k - number of iterations of algorithm
         
     Returns: 
         nextNodes - dictionary to store clusters as "pointers" 
            - the dictionary keys are the nodes and the values are the node in the same cluster that the key node points to
         setStarters - set of nodes that no other node points to (used to construct the sets below)
     
     Algorithm:
         1 sort edges by weight (highest correlation first)
         2 initialize nextNodes so that each node points to itself (single-node clusters)
         3 take highest correlation edge, merge the source and destination nodes' clusters
         4 if max iterations not reached, go back to 3 with next highest correlation
         (meanwhile also keep track of the "set starters" ie nodes that have nothing pointing to them)
    
        
    """
    
    sortedEdges = sorted(edgeList,reverse=True) # sort edges - note reverse!  
    nextNodes = {node:node for node in correl.columns} # dictionary of pointers
    setStarters = {node for node in correl.columns} # keeps track of "set starters", ie nodes that no other node points to in nextNodes 
    
    for i in range(k): # loop until max iterations reached
        weight,src,dest = sortedEdges[i]
        if findBottom(src,nextNodes) != findBottom(dest,nextNodes): # check if already in same cluster
            nextNodes = mergeSets(src,dest,nextNodes,setStarters) # if not, merge clusters
    
    return setStarters,nextNodes


k = 1000
setStarters,nextNodes = kruskalCluster(edges, correl, k)
#%timeit kruskalCluster(edges,k)
```

But for now the algorithm only gives us a representation of the clusters as a dictionary. To make the result more useful, we need to recover the sets themselves in a more intuitive form.

```python
def constructSets(setStarters,nextNodes):
    # Constructs sets (clusters) from the nextNodes dictionary
    # Returns a dictionary of sets
    # Starts from each node in setStarters and loops through until finds a bottom node, adding each node along the way
    # Adds the bottom node as dictionary key (if it does not exist) and the set as the value (or appends to current value)
    allSets = dict()
    for s in setStarters:
        curSet = set()
        curSet.add(s)
        p = s
        while nextNodes[p] != p:
            p = nextNodes[p]
            curSet.add(p)
            
        if p not in allSets:
            allSets[p] = curSet
        else: 
            for item in curSet:
                allSets[p].add(item)
    return allSets

allSets = constructSets(setStarters,nextNodes)


def construct_sets(set_starters, next_nodes):
    """
    Constructs sets (clusters) from the next_nodes dictionary
    
    Parameters:
        set_starters: set of starting nodes 
        next_nodes: dictionary of connections
    
    Returns: 
        dictionary of sets (clusters):
            key - bottom node of set; value - set of all nodes in the cluster
    
    """
    # Initialise an empty dictionary 
    all_sets = dict()
    
    # Loop:
    # Start from each set starter node
    # Create a current set with all nodes on the way to bottom node
    # If bottom node is already in another set, combine the two sets,
    # Otherwis add current set to all_sets
    for s in set_starters:
        cur_set = set()
        cur_set.add(s)
        p = s
        while next_nodes[p] != p:
            p = next_nodes[p]
            cur_set.add(p)
            
        if p not in all_sets:
            all_sets[p] = cur_set
        else: 
            for item in cur_set:
                all_sets[p].add(item)
    return all_sets

allSets = construct_sets(setStarters,nextNodes)

```

Let's print the results in a nicer form.

```python
def getCompanyNames(companyData,symbols):
    # Get company name based on symbol
    return [companyData[s][0] for s in symbols]

def getCompanyIndustry(companyData,symbols):
    # Get company industry based on symbol
    return [companyData[s][1] for s in symbols]

def printSetCompanies(companyData,companySet):
    # Print the names of companies in a set
    names = getCompanyNames(companyData,companySet)
    industries = getCompanyIndustry(companyData,companySet)
    for comp,ind,sym in zip(names, industries, companySet):
        print("{0:10} {1:35} {2}".format(sym,comp,ind)) # format uses {order left to right:length in characters}

def setsIntoList(setdict):
    # Convert dictionary of clusters to list
    allSetsList = []
    for key, value in setdict.items():
        allSetsList.append(value)
    return allSetsList
        
def printLargestClusters(setList,namesDict,n):
    # Print largest clusters
    sortedByLength = sorted(setList,key=len,reverse=True)
    
    for i in range(n):
        s = sortedByLength[i]
        print('Cluster no ' + str(i+1) + '\n')
        printSetCompanies(namesDict,s)
        print('\n')
    return sortedByLength

allSetsList = setsIntoList(allSets) # use a list for convenience
sortedClusters = printLargestClusters(allSetsList,namesDict,20) # print largest clusters
```

Analysing the results, it looks like with $k=300$, only a few large clusters are created, with many smaller clusters. The largest clusters correspond roughly to banks, power companies, real estate, oil, etc. With $k=1000$, these clusters grow significantly, but other clusters also start to emerge, eg consumer staples around Colgate and Kimberly-Clark, airlines, health care, and stock exchanges. The match between industry and cluster is no longer perfect, as some "consumer discretionary" companies like Snap-On (a tool-maker) get lumped together with the banks. Part of the story is also what does not get included in clusters - here, for example, some offshore drilling companies like Transocean are separate from the main oil companies. As you increase the value of $k$ to $4000$, the clusters become larger and arguably much less useful, as they lump all kinds of companies together. A more formal analysis could look at the size and the number of clusters as a function of $k$.

The algorithm is a version of Kruskal's algorithm for finding minimum spanning trees. The MST problem is essentially finding the lowest-cost way of connecting a weighted graph. Here we apply the greedy algorithmic principle used to solve that problem for the problem of clustering: we simply take the "best" edges (here highest correlations) in turn and create connections, but stop short of connecting the entire graph. 

Notice however that this method ignores information about how high the correlations actually are: it merely ranks them in order and uses that information to create the clusters. For clustering purposes, this information would be useful, as now the algorithm makes limited use of the "absolute" distances between firms (consider eg whether average correlation within a cluster is high). This is a significant drawback, and indeed other methods incorporating more information are often preferred.

## Extra part

### In-depth analysis

There are many possibilities here. One is discussing the algorithm and it's implementation and complexity in detail. Our implementation is not most efficient or elegant way to implement Kruskal's algorithm. A full implementation would use a union-find (or disjoint-set) data structure. This could lead to a discussion of tree-type data structures in general, or the complexity of the particular operations we're performing and how to speed them up. 

You may have noticed that the algorithm takes some time to run, especially with large values of $k$. Why is this? As you can see from the pictures in the instructions, the pointers we store in a dictionary essentially form a "tree" of stocks in the same cluster. At every iteration we need to check whether the "source" and "destination" stocks of the current correlation are already in the same cluster/tree. To do this, we hop along the branches of the tree from these stocks all the way to the bottom node and check whether this is the same for the two stocks.  If the branches become long, this may take a relatively long time, possibly up to $O(n)$ depending on the implementation, where $n$ is the number of stocks we have. 

The root(!) of the problem is that we're not being smart about constructing the cluster "tree" - we're simply making one bottom node arbitrarily point to another when merging two clusters. This may lead to the tree depth growing quickly. It turns out we can get a big improvement with a few simple but clever tricks.

Specifically, above we're not using node *ranking* or *path compression*, as explained in the Wikipedia article on the data structure (see https://en.wikipedia.org/wiki/Disjoint-set_data_structure). Keeping track of node ranks allows us to merge clusters such that the depth of the tree grows slowly, and path compression allows us to keep nodes close to the root node.

What is the overall complexity of the algorithm? Sorting $m$ edges is $O(m \log m)$ so we will never do better than that bottleneck if our edges are not already sorted. However, the rest of the algorithm (assuming sorted edges) can be very close to linear (approaching $O(m)$) with the right implementation. Notice that we have $k < m$ and $m = O(n^2)$ here.

Below is a faster version of the algorithm that closely follows the Wikipedia pseudocode. This gives a significant speed gain for large $k$ with relatively small changes to the code.

```python
def findBottom(node,nextNodes):
    # Uses path compression to control tree depth
    if node != nextNodes[node]:
        nextNodes[node] = findBottom(nextNodes[node],nextNodes) # Update pointers towards bottom along the way
    return nextNodes[node]

def mergeSets(node1,node2,nextNodes,ranks):
    # Uses node ranks when merging to control tree depth (union by rank)
    bottom1 = findBottom(node1,nextNodes)
    bottom2 = findBottom(node2,nextNodes)
    # Attach smaller tree to the root of larger tree
    if ranks[bottom1] < ranks[bottom2]:    
        nextNodes[bottom1] = bottom2
    elif ranks[bottom2] < ranks[bottom1]:    
        nextNodes[bottom2] = bottom1
    else: # equal ranks
        nextNodes[bottom2] = bottom1
        ranks[bottom1] = ranks[bottom1] + 1 # Tree becomes larger (an extra "layer" of depth)
    return nextNodes

def kruskalClusterFast(edgeList,k):
    # Faster implementation using rank union and path compression as above
    sortedEdges = sorted(edgeList,reverse=True)    
    
    nextNodes = {node:node for node in correl.columns}
    ranks = {node:0 for node in correl.columns} # Initialize ranks for each node to be zero 
    
    for i in range(k):
        weight,src,dest = sortedEdges[i]
        if findBottom(src,nextNodes) != findBottom(dest,nextNodes):
            nextNodes = mergeSets(src,dest,nextNodes,ranks)
    return nextNodes

def constructSetsSimple(nextNodes):
    # We don't really need setStarters for this implementation - we can just loop through everything quickly as the tree is flat
    allSets = dict()
    for i,s in enumerate(nextNodes):
        curSet = set()
        curSet.add(s)
        p = s
        while nextNodes[p] != p:
            p = nextNodes[p]
            curSet.add(p)
        if p not in allSets:
            allSets[p] = curSet
        else: 
            for item in curSet:
                allSets[p].add(item)
    return allSets


k = 1000
nextNodes = kruskalClusterFast(edges,k)
%timeit kruskalClusterFast(edges,k)

#allSets = constructSetsSimple(nextNodes)
#allSetsList = setsIntoList(allSets) # use a list for convenience
#sortedClusters = printLargestClusters(allSetsList,namesDict,20) # print largest clusters

```

### Other clustering methods

As we have seen, the greedy clustering algorithm we've been using to some extent ignores information that we have. Indeed it is not the most commonly used clustering algorithm. Many other approaches exist; for instance K-means clustering is very common and arguably more intuitive and useful for many clustering applications (although perhaps not here). You could run many of these clustering algorithms from `scikit-learn` using the fairly complete tutorial. Here is an application of one of the algorithms: affinity propagation, which is also given as an example for stock clustering in the tutorial. Affinity propagation attempts to do something a bit more sophisticated than our algorithm: find "typical" nodes that "do well" in representing the other nodes in a cluster based on similarity scores (eg correlations here). It also endogenously decides the best number of clusters based on the data. See eg the Wikipedia page for more information: https://en.wikipedia.org/wiki/Affinity_propagation.

Let's follow the tutorial steps for our data. This results in larger and broader-based clusters than Kruskal's algorithm. Many cluster types are preserved - banks, electricity, oil, etc, but it is interesting to observe how industrials, IT firms, and other companies are divided to groups, and teasing out their connections. Arguably this is a better way to cluster than just using the industry definitions. There's of course a lot more that could be written here on the results and comparing the algorithm and other algorithms. We're still using just correlation data, whereas we could use many other metrics derived from the returns too... notice that for many of the clustering algorithms, you can't simply plug in the correlations as data - you'll need to use other features.


```python
from sklearn import cluster

X = np.array(correl)
_, labels = cluster.affinity_propagation(X)

symbols = np.array(priceData.columns)

def createSetList(symbols,labels):
    # Make output similar as above for printing purposes
    setlist = []
    n_labels = labels.max()
    for i in range(n_labels + 1):
        cluster = set()
        symbs = symbols[labels == i]
        cluster.update(symbs)
        setlist.append(cluster)
    return setlist

setlist = createSetList(symbols,labels)

sortedList = printLargestClusters(setlist,namesDict,15)
```
